```
#How good are you really?
#8 kyu
def better_than_average(class_points, your_points):
  return (sum(class_points)/len(class_points) < your_points)
  
print(better_than_average([50,46,65,70],65))
```

```
#Expressions Matter
#8 kyu
def expression_matter(a, b, c):
    return max(
        a*b*c,
        a+b+c,
        (a+b)*c,
        a+b*c,
        a*(b+c))
    
print(expression_matter(2,4,6))
```

```
#Thinkful - Number Drills: Blue and red marbles
#8 kyu
def guess_blue(blue_start, red_start, blue_pulled, red_pulled):
  return (blue_start-blue_pulled)/((blue_start-blue_pulled)+(red_start-red_pulled))
  
print(guess_blue(10,5,3,4))
```

```
#Grasshopper - Grade book
#8 kyu
def get_grade(s1, s2, s3):
    score=((s1+s2+s3)/3)
    if score >= 90: return "A" 
    if score >= 80: return "B"
    if score >= 70: return "C"
    if score >= 60: return "D" 
    return "F"
    
print(get_grade(40,60,85))
```

```
#No zeros for heros
#8 kuy
def no_boring_zeros(n):
  n = str(n)
  if n != "0":
    while n[-1] == "0": 
      n = n[:-1]
  return int(n)
  
print(no_boring_zeros(1002100))
```

```
#Removing Elements
#8 kuy
def remove_every_other(my_list):
    new_list = []
    for i in range (0, len(my_list),2):
        new_list.append(my_list[i])
    return new_list
    pass
    
print(remove_every_other(['Hello', 'Goodbye', 'Hello Again']))
```

```
#noobCode 01: SUPERSIZE ME.... or rather, this integer!
#8 kyu
def super_size(n):
  nlist = list(str(n))
  nlist.sort(reverse=True)
  newn = ''.join(nlist)
  return int(newn)
  
print(super_size(3456225))
```
